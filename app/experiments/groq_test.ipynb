# Groq + LangChain Exploration

## Objective
This notebook is created to verify the integration of the Groq API with LangChain.

### Goals
- Load environment variables from `.env`
- Initialize the Groq LLM
- Send a prompt using LangChain's `HumanMessage`
- Receive and print the model's response

This notebook is part of the LLM Chatbot Assignment.

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
import os
```

```python
load_dotenv()

llm = ChatGroq(
    model=os.getenv("MODEL_NAME"),
    api_key=os.getenv("GROQ_API_KEY")
)
```

```python
response = llm.invoke(
    [
        HumanMessage(
            content="hey,kaisahai"
        )
    ]
)

print(response.content)
```

## Observations

- Successfully loaded environment variables from the `.env` file.
- Initialized the Groq model using LangChain's `ChatGroq`.
- Sent a prompt using `HumanMessage`.
- Received a valid response from the Groq API.

## Conclusion

The Groq API integration with LangChain is working correctly. This experiment verifies that the chatbot can successfully communicate with the LLM before integrating it into the FastAPI application.
